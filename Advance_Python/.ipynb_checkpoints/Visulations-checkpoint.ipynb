```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
import pandas as pd 

%matplotlib inline
```

```python
# bar chart 

labels = [ "India", "Aus", "NZ", "PAK","AFG"]
freq = [ 260, 290, 220, 190, 60 ]
```

```python
plt.bar(labels,height=freq,color=['r','g','b','m','c'])
plt.show()
```

```python
plt.barh(labels,width=freq,color=['r','g','b','m','c'])
plt.show()
```

```python
# bar chart 
labels = [ "India", "Aus", "NZ", "PAK","AFG"]
freq = [ 260, 290, 220, 190, 60 ]
color = [ 'b','y','k','g','m'] 
e = [ 0.2, 0, 0, 0.1, 0]
plt.pie(x=freq,labels=labels,colors=color,shadow=True,
       autopct="%.2f",counterclock=False,
       startangle=90,explode=e)
plt.show()
```

```python
normal = np.random.normal(5.5,1.5,1000)
```

```python
plt.hist(normal,30,edgecolor='black')
plt.show()
```

```python
plt.boxplot(normal)
plt.show()
```

```python
outlier = np.random.normal(1.5,.5,500)
data  = np.append(normal,outlier)
plt.hist(data,30,edgecolor='black')
plt.plot([data.mean(),data.mean()],[0,190],'r')
plt.plot([np.median(data),np.median(data)],[0,190],'b')
plt.show()
```

```python
# live plots 
import psutil
import time
for var in range(10): 
    print(psutil.cpu_percent(),flush=True)
    time.sleep(2)
```

```python

import matplotlib.pyplot as plt
from psutil import cpu_percent as cpu 
from time import sleep
from matplotlib.animation import FuncAnimation
plt.rcParams['animation.html'] = 'jshtml'
%matplotlib notebook
x = list(range(1,51))
y = []
for var in range(50): 
    y.append(cpu())
    sleep(.2)  
fig,ax = plt.subplots()
ax.plot(x,y,'ro--')
def change(interval):
    x.pop(0)
    y.pop(0)
    x.append(x[-1]+1)
    y.append(cpu())
    ax.clear()
    ax.plot(x,y,'ro--')
    sleep(.5)
ani = FuncAnimation(fig,change,2000)
plt.show()
```
