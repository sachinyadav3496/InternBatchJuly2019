```python
deg = np.arange(0,365,15)
rad = np.deg2rad(deg)

sin = np.sin(rad)
cos = np.cos(rad)
tan = np.tan(rad)
```

```python
plt.figure(figsize=(8,4))
plt.plot(deg,sin,'ro--',label='Sin$\Theta$')
plt.plot(deg,cos,'go--',label='Cos$\Theta$')
#plt.plot(deg,tan,'bo--',label='Tan$\Theta$')
plt.legend(fontsize=15)
plt.xlabel("degree",fontsize=15)
plt.ylabel("values",fontsize=15)
plt.title("Siv Vs Cos Vs Tan",fontsize=20)
plt.xticks(np.arange(0,361,30))
#plt.yticks(np.arange(-1,1,.2))
plt.grid(color='gray')
plt.show()
```

```python
fig,ax  = plt.subplots(facecolor='gray',figsize=(10,4))
x = np.linspace(-5,5,500)
ax.plot(x,x**2,label='$X^2$',lw=3)
ax.plot(x,x**3,label='$X^3$',lw=3)
ax.legend(fontsize=15)
ax.set_facecolor('#123456')
ax.set_xlabel("X axis",fontsize=15)
ax.set_ylabel("Y axis",fontsize=15)
ax.set_title("$X^2$ Vs $X^3$",fontsize=20)
ax.set_xticks(np.arange(-10,10,2))
ax.set_yticks(np.arange(-120,120,20))
plt.savefig("myfig.png")
ax.set_xlim([-20,20])
ax.set_ylim([-150,150])
ax.grid()
plt.show()
```

```python
fig,ax = plt.subplots(nrows=1,ncols=3,figsize=(8,3))
#print(ax)
fig.set_facecolor("#abcdef")
ax[2].plot(x,x**4)
ax[1].plot(x,x**3)
ax[0].plot(x,x**2)
plt.show()
```

```python
fig,ax = plt.subplots(nrows=1,ncols=3,figsize=(8,3))
power = 2
fig.set_facecolor("#abcdef")
title =  [ "$X^2$","$X^3$","$X^4$"]
ax[0].set_ylabel("Y_axis")
for axes in ax : 
    axes.plot(x,x**power)
    axes.spines['right'].set_visible(False)
    axes.spines['top'].set_visible(False)
    axes.set_facecolor("#abcdef")
    axes.set_title(title[power-2])
    axes.set_xlabel("X_axis")
    power += 1
plt.show()
```

```python
fig,ax = plt.subplots(nrows=3,ncols=2,sharex=True,sharey=True)
x = np.linspace(-5,5,100)
power = 2
for axes in ax :  
        axes[0].plot(x,x**power)
        axes[1].plot(x,(x+5)**power)
        power += 1
        axes[0].spines['right'].set_visible(False)
        axes[1].spines['right'].set_visible(False)
        axes[0].spines['top'].set_visible(False)
        axes[1].spines['top'].set_visible(False)
        break
        
```

```python
height = np.random.normal(5.5,1.5,100)
weight = height*20.5 + np.random.random(100)*60
plt.scatter(height,weight)
plt.xlabel("HEIGHT")
plt.ylabel("WEIGHT")
plt.show()
```
