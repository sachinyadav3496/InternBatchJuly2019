```python
fig,ax = plt.subplots(figsize=(10,3))
sns.countplot(diamond['cut'],ax=ax)
height = []
for pat in ax.patches : 
    height.append(pat.get_height())
per = []
s = sum(height)
for h in height : 
    per.append(str(round(h/s,2)*100)+" %")
for pat,p in zip(ax.patches,per) : 
    height = pat.get_height()+1150
    width = pat.get_x()+0.2
    ax.text(width,height,p,fontsize=20)
plt.show()
```
